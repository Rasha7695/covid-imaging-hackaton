```
import pandas as pd
df1 = pd.read_csv('https://raw.githubusercontent.com/chocobearz/covid-imaging-hackaton/master/metadata.csv')
```

```
df1['patientid'].nunique()
```

```
df1s = df1.drop_duplicates(subset="patientid", keep="last")
df1s
```

```
print(df1s['sex'].count())
print(df1s['age'].count())
print(df1s['finding'].count())
```

```
df1s = df1s.dropna(subset=['sex', 'age','survival'])
len(df1s)
```

```
df1s = df1s.loc[df1s['finding'] == 'COVID-19']
df1s
```

```
df_survial_Y = df1s.loc[df1s['survival'] == 'Y']
df_survial_N = df1s.loc[df1s['survival'] == 'N']
```

```
print(df_survial_Y)
print(df_survial_N)
```

```
df_survial_Y.to_csv('df_survial_Y.csv')
df_survial_N.to_csv('df_survial_N.csv')
```
